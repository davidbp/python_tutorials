### Get top K values sorted by date in duckdb

```python
from typing import Iterable, Optional

def get_top_k_rows_sorting_by_date(
    file_path: Path,
    order_col: str,
    nrows: Optional[int] = None
):
    """
    

    If `nrows` is provided the method gets the top `nrow` asins from the `col_values`
    This function avoids loading all the dataset in memory.
    """

    query_str = f"""SELECT *
                    FROM "{file_path}"
                    ORDER BY "{order_col}"
                 """
    if nrows:
        query_str += f"LIMIT {nrows}\n"

    df = duckdb.query(query_str).df()
    return df
```

```python
df_res = get_top_k_rows_sorting_by_date(file_path, order_col="reviewTime")
```

### Get top K rows with unique values from col_unique sorted by col_date in duckdb

```python
from typing import Iterable, Optional

def get_top_k_rows_sorting_by_date2(
    file_path: Path,
    order_col: str,
    nrows: Optional[int] = None
):
    """
    

    If `nrows` is provided the method gets the top `nrow` asins from the `col_values`
    This function avoids loading all the dataset in memory.
    """

    query_str = f"""SELECT 
                        a.asin, 
                        a.summary, 
                        a.reviewTime
                    FROM "{file_path}" a
                    INNER JOIN (
                        SELECT asin, MIN(reviewTime) AS min_reviewTime
                        FROM "{file_path}" 
                        GROUP BY asin
                    ) b
                        ON a.asin = b.asin
                        AND a.reviewTime = b.min_reviewTime
                    ORDER BY a.reviewTime ASC;
                 """
    if nrows:
        query_str += f"LIMIT {nrows}\n"

    df = duckdb.query(query_str).df()
    return df
```

```python
from typing import Iterable, Optional

def get_asins_if_col_matches_values(
    file_path: Path, retrieved_cols:Iterable[str], col: str, col_values: Iterable[str], nrows: Optional[int] = None
):
    """
    Get the `retrieved_cols` of a given col from a dataset of (asin, col) pairs.

    If `nrows` is provided the method gets the top `nrow` asins from the `col_values`
    This function avoids loading all the dataset in memory.
    """

    col_values_str_set = f"{tuple(x for x in col_values)}"
    query_str = f"""SELECT *
                    FROM "{file_path}"
                    WHERE {col} IN {col_values_str_set}
                 """
    if nrows:
        query_str += f"LIMIT {nrows}\n"

    df = duckdb.query(query_str).df()
    return df
```

```python
from typing import Iterable, Optional

def get_asins_if_col_matches_values(
    file_path: Path, retrieved_cols:Iterable[str], col: str, col_values: Iterable[str], nrows: Optional[int] = None
):
    """
    Get the `retrieved_cols` of a given col from a dataset of (asin, col) pairs.

    If `nrows` is provided the method gets the top `nrow` asins from the `col_values`
    This function avoids loading all the dataset in memory.
    """

    col_values_str_set = f"{tuple(x for x in col_values)}"
    retrieved_cols_set =  ','.join(retrieved_cols)
    query_str = f"""SELECT {retrieved_cols_set}
                    FROM "{file_path}"
                    WHERE {col} IN {col_values_str_set}
                 """
    if nrows:
        query_str += f"LIMIT {nrows}\n"

    df = duckdb.query(query_str).df()
    return df
```
