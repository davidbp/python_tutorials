# Plotting Lines

It is quite standard to import `matplotlib.pyplot` as `plt`.

The basic function to make a line is the `plt.plot` function which takes two arguments `x` and `y`.

- `plt.plot(x,y, linestyle, alpha, color)` will make a plot at the X axis position `x[k]` with height `y[k]` with style `style` and color with transparency `alpha` and color `color`.

#### About linestyle

The style is a string describing how points will be plotted. Most styles can be summarized as with two characters.

- Characters in `[o,x,s,d,p,h,v,+]` define how the dot at position `(x[k],y[k])` is plotted

- Characters in  `[-,:,.]` define how the space between two points is plotted

#### About transparency

The transparency is a float value between 0(transparent) and 1(opaque).

#### About color

Colors can be specified as a string containing the name of the color. Some possible colors `["blue", "red", "violet", "black", "green", "yellow"]`.



#### Some examples

- `linestyle="-"` will use regular lines


- `linestyle="o"` will mark $(x_k,y_k)$ with dots (no conection between dots, therefore no lines)


- `linestyle="o-"` will mark $(x_k,y_k)$ with circles. Circles will be linked by lines


- `linestyle="--"` will use dashed lines.


- `linestyle="-."` will use dashed lines followed by dots.

### Display error on the y axis

Many times we want to show errors on the measures of the Y axis.

We can do this using in two different ways. We can use `plt.errobar` or  `plt.fill_between`


- `plt.errorbar(x , y, error)` makes a vertical line containing the error specified by `error` at every position  `(x[k], y[k])`.


- `plt.fill_between(x , y-error, y+error)` makes a shaded region painted between `(x[k], y[k]-error[k]` and `(x[k], y[k]+error[k])`

```python
x = np.linspace(0, 30, 30)
y = np.sin(x/6*np.pi)
error = np.random.normal(0.1, 0.02, size=y.shape)*5
y    += np.random.normal(0, 0.1, size=y.shape)

plt.plot(x, y, '--',alpha=0.5, color="blue")
plt.errorbar(x,y,error)
plt.title("Sinus samples with the expected error seen shaded")
plt.show()
```

```python

x = np.linspace(0, 30, 30)
y = np.sin(x/6*np.pi)
error = np.random.normal(0.1, 0.02, size=y.shape)*5
y    += np.random.normal(0, 0.1, size=y.shape)

plt.plot(x, y, '--',alpha=0.5)
plt.fill_between(x, y-error, y+error,alpha=0.2)
plt.title("Sinus samples with the expected error seen shaded")
plt.show()
```

#### Typical scenario for error lines

Imagine some process that has noise. Here we have 3 measures of `y` in `[y1,y2,y3]`, we can use the function `fill_beteween` to plot the overall mean of the different components and the standard deviation in a single plot. 

This plot will make a summary of the "mean of the trajectory" of whatever we are plotting.

```python
x = np.linspace(0, 30, 30)
y = np.sin(x/6*np.pi)

y1  = y +  np.random.normal(0, 0.8, size=y.shape)
y2  = y +  np.random.normal(0, 0.8, size=y.shape)
y3  = y +  np.random.normal(0, 0.8, size=y.shape)

y_mean   = np.mean([y1,y2,y3],axis=0)
y_std = np.std([y1,y2,y3],axis=0)


plt.plot(x, y_mean, '--',alpha=0.5)
plt.fill_between(x, y_mean-y_std, y_mean+y_std,alpha=0.2)
plt.title("Sinus samples with the expected error seen shaded")
plt.show()
```
