## Superpose plots

We can superpose two plots specifying the same axis in different `.plot` calls.

Note that it is imporatnt to put a low `alpha` value in one of the two plots that we superpose, otherwise
one of the superposed plots will mask the other. In the following example the first subplot has no alpha value
hence the bars hide the curve. The second subplot has alpha=0.2 which allows the curve to be seen.

```python
date1 = pd.datetime(2018, 4, 10)
data = {'size': [314, 253, 282, 292],
        'height': [34,123,4,1],
        'width': [20,40,50,10]}

df = pd.DataFrame(data)

fig, ax1 = plt.subplots(nrows=2,figsize = (10, 10))

df[['size']].plot(kind = 'bar', stacked = True, ax = ax1[0])
df['height'].plot(ax = ax1[0])

df[['size']].plot(kind = 'bar', stacked = True, ax = ax1[1], alpha=0.2)
df['width'].plot(ax = ax1[1])

plt.show()
```

Moreover we can share an axis using the `plt.twinx` function.
Note that the second subplot in this example plots in  `ax2 = plt.twinx(ax1[1])`.
This essentially creates a new Y axis and scales the plot in the new Y axis whosh in the right hand side.


```python
date1 = pd.datetime(2018, 4, 10)
data = {'size': [314, 253, 282, 292],
        'height': [34,123,4,1],
        'width': [20,40,50,10]}

df = pd.DataFrame(data)

fig, ax1 = plt.subplots(nrows=2,figsize = (10, 10))

df[['size']].plot(kind = 'bar', stacked = True, ax = ax1[0])
df['height'].plot(ax = ax1[0])

df[['size']].plot(kind = 'bar', stacked = True, ax = ax1[1], alpha=0.2)
ax2 = plt.twinx(ax1[1])
df['width'].plot(ax = ax2)

plt.show()
```
